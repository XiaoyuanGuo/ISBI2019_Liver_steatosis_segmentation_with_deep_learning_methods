## Observe_relationship_steatosis_features
is used to analyze the relationship of different properities of the obtained steatosis segmentation, including the eccentricity, area, perimeter and prediction socre of different steatosis segmentation instances.
The input is a .csv file that can be obtained in the Filter_prediction_results.py file.

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
import pandas as pd 
import csv
```

```python
filename = 'feature_analysis'
data = pd.read_csv("{}.csv".format(filename)) 
```

```python
data.head()
```

```python
eccentricity=data['eccentricity']
area = data['area']
perimeter = data['perimeter']
score = data['score']
```

```python
np.mean(area),min(area),max(area),min(area)/np.mean(area),max(area)/np.mean(area)
```

```python
np.mean(eccentricity),min(eccentricity),max(eccentricity),min(eccentricity)/np.mean(eccentricity),max(eccentricity)/np.mean(eccentricity)
```

```python
np.mean(perimeter),min(perimeter),max(perimeter),min(perimeter)/np.mean(perimeter),max(perimeter)/np.mean(perimeter)
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(style="white")

# Generate a large random dataset
#rs = np.random.RandomState(33)
#d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                 #columns=list(ascii_letters[26:]))

# Compute the correlation matrix
corr = data.corr()
# Generate a mask for the upper triangle
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
# Drop non-numerical variables
df=data
numerical = list(set(df.columns) )
# Calculate and plot
corr_matrix = df[numerical].corr()
sns.heatmap(corr_matrix);
plt.savefig('Analyze.png', dpi=400, bbox_inches='tight',pad_inches=0)
```
